```python
# Find which country names in df_pov_indx are not in df_TB_burden
missing_countries = df_pov_indx[df_pov_indx["country_name"].isin(df_TB_burden["country_name"])]["country_name"].unique()

# Check if all country names are present
all_countries_present = len(missing_countries) == 0

# Print results
print("All country_name values from df_pov_indx are in df_TB_burden:", all_countries_present)

if not all_countries_present:
    print("Missing country_name values:", missing_countries)
```

```python
# Adding the iso2 (two letter country code) and iso3 (three letter country code) to the df_pov_indx dataframe
!pip install pycountry
import pycountry
```

```python
# Function to get ISO2 and ISO3 codes from country name
def get_country_codes(country_name):
    try:
        country = pycountry.countries.lookup(country_name)
        return country.alpha_2, country.alpha_3
    except LookupError:
        return None, None  # Return None if country is not found

# Apply function to add ISO codes to df_pov_indx
df_pov_indx["iso2"], df_pov_indx["iso3"] = zip(*df_pov_indx["country_name"].apply(get_country_codes))

# Display updated dataframe
print(df_pov_indx.head())
```

```python
# Add rifampicin resistance rates from WHO for 2023 stored in df_RRR to df_TB_burden

# Select specific columns from df_RRR (including "iso3" for merging)
columns_to_add = ["iso3", "source_rr_new", "e_rr_pct_new", "e_inc_rr_num"]
df_RRR_selected = df_RRR[columns_to_add]

# Merge df_TB_burden with df_RRR on "iso3" - left join to keep all rows of df_TB_burden
df_TB_burden_enriched1 = df_TB_burden.merge(df_RRR_selected, on="iso3", how="left")

# Check for unmatched "iso3" values in df_RRR
unmatched_iso3 = df_RRR_selected[~df_RRR_selected["iso3"].isin(df_TB_burden["iso3"])]["iso3"].unique()

# Print error message if there are unmatched rows
if len(unmatched_iso3) > 0:
    print("Error: The following iso3 values in df_RRR were not found in df_TB_burden:", unmatched_iso3)

# Display the merged dataframe
print(df_TB_burden_enriched1.head())
```

```python
# Add BCG vaccination coverage rates from WHO for 2023 stored in df_BCG to df_TB_burden

# Select specific columns from df_BCG (including "iso3" for merging)
columns_to_add = ["iso3", "vaccine_antigen", "coverage_category", "targets_number", "doses", "BCG_coverage"]
df_BCG_selected = df_BCG[columns_to_add]

# Merge df_TB_burden with df_BCG on "iso3" - left join to keep all rows of df_TB_burden
df_TB_burden_enriched2 = df_TB_burden_enriched1.merge(df_BCG_selected, on="iso3", how="left")

# Check for unmatched "iso3" values in df_BCG
unmatched_iso3 = df_BCG_selected[~df_BCG_selected["iso3"].isin(df_TB_burden_enriched1["iso3"])]["iso3"].unique()

# Print error message if there are unmatched rows
if len(unmatched_iso3) > 0:
    print("Error: The following iso3 values in df_BCG were not found in df_TB_burden_enriched1:", unmatched_iso3)

# Display the merged dataframe
print(df_TB_burden_enriched2.head())

# Caution: Kosovo (iso3 = XKX) in df_BCG is dropped, because it is not present in df_TB_burden
```

```python
# Add population density per country for 2023 (people/km²) from Earth Database stored in df_pop_dens_iso to df_TB_burden

# Select specific columns from df_pop_dens_iso (including "iso3" for merging)
columns_to_add = ["iso3", "population_density"]
df_pop_dens_iso_selected = df_pop_dens_iso[columns_to_add]

# Merge df_TB_burden with df_pop_dens_iso on "iso3" - left join to keep all rows of df_TB_burden
df_TB_burden_enriched3 = df_TB_burden_enriched2.merge(df_pop_dens_iso_selected, on="iso3", how="left")

# Check for unmatched "iso3" values in df_pop_dens_iso
unmatched_iso3 = df_pop_dens_iso_selected[~df_pop_dens_iso_selected["iso3"].isin(df_TB_burden_enriched2["iso3"])]["iso3"].unique()

# Print error message if there are unmatched rows
if len(unmatched_iso3) > 0:
    print("Error: The following iso3 values in df_pop_dens_iso were not found in df_TB_burden_enriched2:", unmatched_iso3)

# Display the merged dataframe
print(df_TB_burden_enriched3.head())

# SEVERAL COUNTRIES WERE DROPPED: 'GIB' 'VAT' 'GGY' 'JEY' 'MYT' 'TWN' 'MAF' 'BLM' 'REU' 'MTQ' 'LIE' 'UMI' 'GLP' 'XXK' 'IMN' 'BES' 'FRO' 'SPM' 'SHN' 'GUF' 'ESH' 'FLK'
```

```python
# Add poverty index per country for 2023 from United Nations Human Development Report stored in df_pov_indx_iso to df_TB_burden

# Select specific columns from df_pov_indx_iso (including "iso3" for merging)
columns_to_add = ["iso3", "MPI_value", "pop_in_MP_percent"]
df_pov_indx_iso_selected = df_pov_indx_iso[columns_to_add]

# Merge df_TB_burden with df_pov_indx_iso on "iso3" - left join to keep all rows of df_TB_burden
df_TB_burden_enriched4 = df_TB_burden_enriched3.merge(df_pov_indx_iso_selected, on="iso3", how="left")

# Check for unmatched "iso3" values in df_pov_indx_iso
unmatched_iso3 = df_pov_indx_iso_selected[~df_pov_indx_iso_selected["iso3"].isin(df_TB_burden_enriched3["iso3"])]["iso3"].unique()

# Print error message if there are unmatched rows
if len(unmatched_iso3) > 0:
    print("Error: The following iso3 values in df_pov_indx_iso were not found in df_TB_burden_enriched3:", unmatched_iso3)

# Display the merged dataframe
print(df_TB_burden_enriched4.head())
```

```python
# Add smoking rates (%) per country for 2022 form World Population Review stored in df_smoking_iso to df_TB_burden

# Select specific columns from df_smoking_iso (including "iso3" for merging)
columns_to_add = ["iso3", "total_smokers_2022_percent"]
df_smoking_iso_selected = df_smoking_iso[columns_to_add]

# Merge df_TB_burden with df_smoking_iso on "iso3" - left join to keep all rows of df_TB_burden
df_TB_burden_enriched5 = df_TB_burden_enriched4.merge(df_smoking_iso_selected, on="iso3", how="left")

# Check for unmatched "iso3" values in df_smoking_iso
unmatched_iso3 = df_smoking_iso_selected[~df_smoking_iso_selected["iso3"].isin(df_TB_burden_enriched4["iso3"])]["iso3"].unique()

# Print error message if there are unmatched rows
if len(unmatched_iso3) > 0:
    print("Error: The following iso3 values in df_smoking_iso were not found in df_TB_burden_enriched4:", unmatched_iso3)

# Display the merged dataframe
print(df_TB_burden_enriched5.head())
```

```python
# Add air pollution data (annual average exposure to particulate matter < 2.5µm (PM2.5) concentration in μg/m³)
# for 2023 from IQAIR stored in df_air_pollution_iso to df_TB_burden

# Select specific columns from df_air_pollution_iso (including "iso3" for merging)
columns_to_add = ["iso3", "avg_air_pollution_PM2-5_in_2023"]
df_air_pollution_iso_selected = df_air_pollution_iso[columns_to_add]

# Merge df_TB_burden with df_air_pollution_iso on "iso3" - left join to keep all rows of df_TB_burden
df_TB_burden_enriched6 = df_TB_burden_enriched5.merge(df_air_pollution_iso_selected, on="iso3", how="left")

# Check for unmatched "iso3" values in df_air_pollution_iso
unmatched_iso3 = df_air_pollution_iso_selected[~df_air_pollution_iso_selected["iso3"].isin(df_TB_burden_enriched5["iso3"])]["iso3"].unique()

# Print error message if there are unmatched rows
if len(unmatched_iso3) > 0:
    print("Error: The following iso3 values in df_air_pollution_iso were not found in df_TB_burden_enriched5:", unmatched_iso3)

# Display the merged dataframe
print(df_TB_burden_enriched6.head())
```
