```python
# import libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load WHO TB Burden - incidence number 2023 enriched with external data 
TB_burden_enriched = "df_TB_burden_enriched.csv"
df_TB_burden_enriched = pd.read_csv(TB_burden_enriched)
df_TB_burden_enriched
```

```python
# Summary of the data: checking data types
column_summary_df_TB_burden_enriched = pd.DataFrame({
    'Column Name': df_TB_burden_enriched.columns,
    'Data Type': df_TB_burden_enriched.dtypes.values,
    'Non-Null Count': df_TB_burden_enriched.notnull().sum().values,
    'Null Count': df_TB_burden_enriched.isnull().sum().values,
    'Data Type': df_TB_burden_enriched.dtypes.values,
    'Percentage Null': df_TB_burden_enriched.isnull().mean().values * 100,
    'Unique Values': df_TB_burden_enriched.nunique().values
})

print(column_summary_df_TB_burden_enriched)
```

```python
# Calculate TB incidence rate in percent and add it to the dataframe:
df_TB_burden_enriched["e_tb_inc_prct"] = (df_TB_burden_enriched["e_inc_num"] / df_TB_burden_enriched["e_pop_num"]) * 100
df_TB_burden_enriched.head(20)
```

```python
# Summary of the data: checking numerical distribution of data
df_TB_burden_enriched.describe()
```

```python
# Display the 25 longest "country_names" in the dataframe "df_TB_burden_enriched"
longest_countries = (
    df_TB_burden_enriched.assign(name_length=df_TB_burden_enriched["country_name"].str.len())
    .sort_values(by="name_length", ascending=False)
    .head(25)
)

# Display the result
print(longest_countries[["country_name", "name_length"]])
```

```python
# List of shorter names to replace the longest country names
shorter_names = [
    "Palestinian territory", "UK of Great Britain + Northern Ireland", "North Korea", "Venezuela", "Congo", 
    "Bolivia", "Micronesia", "Saint Vincent + Grenadines", "Lao", "Netherlands", "Tanzania", "Iran"
]

# Replace the 12 longest country names with shorter ones
for i, country in enumerate(longest_countries["country_name"].head(12)):
    df_TB_burden_enriched.loc[df_TB_burden_enriched["country_name"] == country, "country_name"] = shorter_names[i]

# Check the modified DataFrame
print(df_TB_burden_enriched[["country_name"]].head(50))  # Display updated country names
```

```python
# Caution: there is four entries for Congo, which need to reduced into one value.
# The first entered row is actually correct, the other three need to be dropped.

# Identify the first occurrence of 'Congo' and keep it, drop the others
df_TB_burden_enriched = df_TB_burden_enriched.drop_duplicates(subset="country_name", keep="first")

# Check the result
print(df_TB_burden_enriched[df_TB_burden_enriched["country_name"] == "Congo"])
```

```python
# The second row was kept instead of the first one.  :-/
# Values for "MPI_value" and "pop_in_MP_percent" will be changed for the country Congo manually:

df_TB_burden_enriched.loc[df_TB_burden_enriched["country_name"] == "Congo", "MPI_value"] = 0.331188736
df_TB_burden_enriched.loc[df_TB_burden_enriched["country_name"] == "Congo", "pop_in_MP_percent"] = 64.51797964

# Check the result for Congo
print(df_TB_burden_enriched[df_TB_burden_enriched["country_name"] == "Congo"][["country_name", "MPI_value", "pop_in_MP_percent"]])
```

```python
# Categorization of the tuberculosis incidence rate (%) in the column "e_tb_inc_prct" into tuberculosis severity levels.
# Based on SD intervals: 0.1 (Mean), 0.244 (Mean to Mean + 1 SD), & 0.388 (Mean + 1 SD to Mean + 2 SD) --> 5 Levels

def categorize_tb_severity(value):
    if value <= 0.05:
        return "Very Low"
    elif value <= 0.1:
        return "Low"
    elif value <= 0.244:
        return "Moderate"
    elif value <= 0.388:
        return "High"
    else:
        return "Critical"

df_TB_burden_enriched["tb_severity"] = df_TB_burden_enriched["e_tb_inc_prct"].apply(categorize_tb_severity)

print(df_TB_burden_enriched[["country_name", "e_tb_inc_prct", "tb_severity"]].head())
```

```python
# Visualization: Display top 5 countries per world region according to: TB incidences

# Group by 'g_whoregion' and select the top 5 countries per region based on 'e_tb_inc_prct'
top_countries_TBinc = (
    df_TB_burden_enriched
    .sort_values(['g_whoregion', 'e_tb_inc_prct'], ascending=[True, False])
    .groupby('g_whoregion')
    .head(5)
)

# Define custom colors for WHO regions
region_colors = {
    "AFR": "#000000",  # Black
    "AMR": "#d62728",  # Red
    "EMR": "#2ca02c",  # Green
    "EUR": "#1f77b4",  # Blue
    "SEA": "#ff7f0e",  # Orange
    "WPR": "#8c564b"   # Brown
}

# Plot the barplot
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    data=top_countries_TBinc, 
    x='country_name', 
    y='e_tb_inc_prct', 
    hue='g_whoregion', 
    palette=region_colors,  # Apply custom colors
    dodge=False
)

# Formatting the plot
plt.xticks(rotation=45, ha='right')
plt.xlabel("Country Name", fontsize=14)
plt.ylabel("TB Incidence in 2023 (%)", fontsize=14)
plt.title("Top 5 Countries with Highest TB Incidence per WHO Region", fontsize=16)
plt.legend(title="WHO Region")
plt.tight_layout()

# Add value labels on top of bars - seems to not work
for bar in ax.patches:  # Get bars from the seaborn plot
    yval = bar.get_height()  # Get bar height (accuracy value)
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:.1f}%", 
             ha='center', va='bottom', fontsize=8, fontweight='bold')

# Show the plot
plt.show()
```

```python
# Create a Pie Chart of TB_severity

# Define custom colors for WHO regions
region_colors = {
    "AFR": "#000000",  # Black
    "AMR": "#d62728",  # Red
    "EMR": "#2ca02c",  # Green
    "EUR": "#1f77b4",  # Blue
    "SEA": "#ff7f0e",  # Orange
    "WPR": "#8c564b"   # Brown
}

# Define color mapping for TB severity levels
severity_colors = {
    "Very Low": "#88CCEE",      # light blue
    "Low": "#44AA99",           # turquise
    "Moderate": "#DDCC77",      # yellow
    "High": "#ff7f0e",          # orange
    "Critical": "#CC6677"       # red
}

# Define the order of TB severity levels (for consistent clockwise arrangement)
severity_order = ["Very Low", "Low", "Moderate", "High", "Critical"]

# Sort WHO regions alphabetically
regions_sorted = sorted(df_TB_burden_enriched["g_whoregion"].unique())

# Create subplots: One pie chart per WHO region
fig, axes = plt.subplots(2, 3, figsize=(18, 12))  # 2 rows, 3 columns for 6 regions
axes = axes.flatten()  # Flatten for easy iteration

# Loop through WHO regions in alphabetical order and create a pie chart for each
for i, region in enumerate(regions_sorted):
    # Filter data for the current region
    region_data = df_TB_burden_enriched[df_TB_burden_enriched["g_whoregion"] == region]

    # Count occurrences of TB severity levels and ensure correct order
    severity_counts = region_data["tb_severity"].value_counts().reindex(severity_order, fill_value=0)

    # Create a pie chart with a clockwise start and "Very Low" at 12 o'clock
    wedges, texts, autotexts = axes[i].pie(
        severity_counts, 
        labels=severity_counts.index, 
        autopct='%1.1f%%', 
        colors=[severity_colors[sev] for sev in severity_counts.index], 
        startangle=90,  # Start "Very Low" at 12 o’clock
        counterclock=False,  # Make it clockwise
        wedgeprops={'edgecolor': region_colors[region], 'linewidth': 2}  # Colored frame
    )

    # Rotate labels to 10 degrees for better readability and set fontsize for labels outside the pie
    for text in texts:
        text.set_fontsize(16)
        text.set_rotation(10)
    
    for autotext in autotexts:  # Percentage values inside the pie
        autotext.set_fontsize(14)
        autotext.set_color("black")  # Ensure readability
        
    # Set title with region name and color it accordingly
    axes[i].set_title(f"WHO Region: {region}", fontsize=16, fontweight="bold", color=region_colors[region])

# Adjust layout to avoid overlap and display the plots
plt.tight_layout()
plt.show()
```
